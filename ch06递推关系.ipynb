##### （3）求$S_n (n\ge 3)$的所有奇子集的容量之和。
记$t(n)$为$S_n$所有子集的容量之和。则：  
$t(n+1) = t(n) + (n+1)|S| = t(n) = 2t(n) + (n+1)2^n, t(3)=24$  
因为当$n\ge 3$时，$S_n$的所有奇子集的容量之和和所有偶子集的容量之和相等，所以$S_n (n\ge 3)$的所有奇子集的容量之和为：$\frac{t(n)}{2}$
